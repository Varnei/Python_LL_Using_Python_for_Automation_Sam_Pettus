```python
Arquivo_Entrada = open('Arquivo_Entrada.txt', 'r')
Arquivo_Passaram = open('Arquivo_Passaram.txt', 'w')
Arquivo_Falharam = open("Arquivo_Falharam.txt", 'w')

Contador_Passaram = 0
Contador_Falharam = 0
Contador_Total = 0

for Linha in Arquivo_Entrada:
    Linha_Dividida = Linha.split()
    
    if Linha_Dividida[2] == 'P':
        Contador_Total = Contador_Total + 1
        Contador_Passaram = Contador_Passaram + 1
        print(str(Contador_Total) + "   ==>   " + Linha_Dividida[0] + ', ' + Linha_Dividida[1] + ' anos, '+ 'passou')
        Arquivo_Passaram.write(Linha)
    
    elif Linha_Dividida[2] =='F':
        Contador_Total = Contador_Total + 1
        Contador_Falharam = Contador_Falharam + 1
        print(str(Contador_Total) + "   ==>   " + Linha_Dividida[0] + ', ' + Linha_Dividida[1] + ' anos, '+ 'falhou')
        Arquivo_Falharam.write(Linha)
        
print ('\nTotal de alunos da classe: ' + str(Contador_Total))
print ('Total de alunos que passaram: ' + str(Contador_Passaram))
print ('Total de alunos que falharam: ' + str(Contador_Falharam))

print ('\nArquivo de saída gerado 1   ==>   Arquivo_Passaram.txt')
print ('Arquivo de saída gerado 2   ==>   Arquivo_Falharam.txt')

Arquivo_Entrada.close()
Arquivo_Passaram.close()
Arquivo_Falharam.close()
```
