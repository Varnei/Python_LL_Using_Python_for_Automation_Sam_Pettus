# Aula 01-02 - Como Ler Arquivos

```python
Arquivo_Entrada = open('Arquivo_Entrada.txt', 'r')

print(Arquivo_Entrada.read())

Arquivo_Entrada.close()
```

```python
Arquivo_Entrada = open('Arquivo_Entrada.txt', 'r')

Contador = 0

for Linha in Arquivo_Entrada:
    Contador = Contador + 1
    print(str(Contador) + " ==> " + Linha)

print ('\nTotal de alunos: ' + str(Contador))

Arquivo_Entrada.close()
```

```python
Arquivo_Entrada = open('Arquivo_Entrada.txt', 'r')

Contador = 0

for Linha in Arquivo_Entrada:
    Linha_Dividida = Linha.split()
    
    if Linha_Dividida[2] == 'P':
        Contador = Contador + 1
        print(str(Contador) + " ==> " + Linha_Dividida[0] + ', ' + Linha_Dividida[1] + ' anos, '+ 'passou')

print ('\nTotal de alunos que passaram: ' + str(Contador))

Arquivo_Entrada.close()
```
