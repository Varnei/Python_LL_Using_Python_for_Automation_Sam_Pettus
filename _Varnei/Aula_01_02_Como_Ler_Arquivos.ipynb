# Aula 01-02 - Como Ler Arquivos

```python
f = open('Arquivo_Entrada.txt', 'r')
print(f.read())
f.close()
```

```python
f = open('Arquivo_Entrada.txt', 'r')
contador = 1
for linha in f:
    print(str(contador) + " ==> " + linha)
    contador = contador + 1
print ('\nTotal de alunos: ' + str(contador - 1))
f.close()
```

```python
f = open('Arquivo_Entrada.txt', 'r')
contador = 1
for linha in f:
    linha_dividida = linha.split()
    if linha_dividida[2] == 'P':
        print(str(contador) + " ==> " + linha_dividida[0] + ', ' + linha_dividida[1] + ' anos, '+ 'Passou')
        contador = contador + 1
print ('\nTotal de alunos que passaram: ' + str(contador - 1))
f.close()
```
